```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars=pd.read_csv('Car details.csv')
cars.head()
```

```python
cars.shape
```

```python
cars.isnull().sum()
```

```python
cars=cars.dropna()
cars.shape
```

```python
import re
```

```python
torque_list=cars['torque'].to_list()
torque_rpm=[]
def extract_rpm(x):
    for item in x:
        residue=item.replace('.','')
        residue=residue.replace(',','')
        temporary=[int(s) for s in re.findall(r'\d+',residue)]
        torque_rpm.append(max(temporary))

extract_rpm(torque_list)
print(torque_list[:2])
print(torque_rpm[:2])
```

```python
cars['torque_rpm']=torque_rpm
cars.head(20)
```

```python
mil_list=cars['mileage'].to_list()
mil_kmpl=[]
def extract_mil(x):
    for item in x:
        temp=[]
        try:
            for s in item.split(" "):
                temp.append(float(s))
        except:
            pass
            mil_kmpl.append(max(temp))
extract_mil(mil_list)
print(mil_list[:2])
print(mil_kmpl[:2])
```

```python
cars['mil_kmpl']=mil_kmpl
cars.head()
```

```python
engine_list=cars['engine'].to_list()
engine_cc=[]
def extract_engine(x):
    for item in x:
        temp=[]
        try:
            for s in item.split(" "):
                temp.append(float(s))
        except:
            pass
            engine_cc.append(max(temp))
extract_engine(engine_list)
print(engine_list[:2])
print(engine_cc[:2])
```

```python
cars['engine_cc']=engine_cc
cars.head()
```

```python
max_power_list=cars['max_power'].to_list()
max_power_bhp=[]
def extract_power(x):
    for item in x:
        temp=[]
        try:
            for s in item.split(" "):
                temp.append(float(s))
        except:
            pass
            max_power_bhp.append(max(temp))
extract_power(max_power_list)
print(max_power_list[:2])
print(max_power_bhp[:2])
```

```python
cars['max_power_bhp']=max_power_bhp
cars.head()
```

```python
cars_new=cars.drop(['mileage','engine','max_power','torque'],axis=1)
cars_new.head()
```

```python
cars_new.shape
```

```python
cars_new.describe()
```

```python
cars_new['fuel'].value_counts()
```

```python
cars_new['seller_type'].value_counts()
```

```python
cars_new['transmission'].value_counts()
```

```python
cars_new['owner'].value_counts()
```

```python
cars_new['transmission']=cars_new['transmission'].map({'Manual':1,'Automatic':0})
```

```python
cars_new['seller_type']=cars_new['seller_type'].map({'Individual':1,'Dealer':0,'Trustmark Dealer':-1})
cars_new.head()
```

```python
cars_new['fuel']=cars_new['fuel'].map({'Petrol':1,'Diesel':0,'CNG':-1,'LPG':-1})
cars_new.head()
```

```python
owners=pd.get_dummies(cars_new['owner'])
owners=owners.replace(True,1)
owners=owners.replace(False,0)
owners.head()
```

```python
X=pd.concat([cars_new,owners],axis=1)
X.head()
```

```python
X=X.drop(['name','owner'],axis=1)
X.head()
```

```python
Y=X['selling_price']
X=X.drop(['selling_price'],axis=1)
```

```python
Y.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
random_model=RandomForestRegressor(n_estimators=300,random_state=42,n_jobs=-1)
```

```python
random_model.fit(x_train,y_train)

random_model_accuracy_train = round(random_model.score(x_train,y_train)*100,2)
random_model_accuracy_train
```

```python
random_model_accuracy_test = round(random_model.score(x_test,y_test)*100,2)
random_model_accuracy_test
```

```python
import pickle
saved_model=pickle.dump(random_model,open('car_price_prediction.pkl','wb'))
```
